```python
new_d=data.drop(index=[1,27])
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x='Murder',y='Assault',data=new_d)
plt.show()
```

```python
X=new_d[['Murder','Assault','UrbanPop','Rape']]
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=2)
```

```python
km.fit(X)
```

```python
clusters=km.predict(X)
```

```python
clusters
```

```python
pd.Series(clusters).value_counts()
```

```python
new_d['clusters']=clusters
```

```python
new_d
```

```python
km.inertia_ 
```

```python
all_wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(X)
    wcss = km.inertia_
    all_wcss.append(wcss)
```

```python
all_wcss
```

```python
plt.plot(range(1,11), all_wcss)
plt.show()
```

```python
km = KMeans(n_clusters=3)
km.fit(X)
clusters = km.predict(X)
```

```python
new_d['clusters']=clusters
```

```python
new_d
```

```python
sns.scatterplot(x='Murder',y='Assault',data=new_d,hue="clusters")
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
mergings = linkage(X, method = "complete")
```

```python
plt.figure(figsize=(10,5))
dendrogram(mergings)
plt.show()
```

```python
labels = fcluster(mergings, 85, criterion='distance')
```

```python
labels
```

```python
new_d['hc_clusters']=labels
```
